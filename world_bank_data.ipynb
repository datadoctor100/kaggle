```python
# Visualize GPD data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.plot(column = 'pop_est', figsize = [50, 20], legend = True)
plt.axis('off'),
plt.title('World Map Based On Estimated Population');
```

```python
# Subset desired columns
keep = ['Country Name', 'Country Code', '2019']
expdat = exports_['data'][[i for i in keep]]
impdat = imports_['data'][[i for i in keep]].rename({'2019': 'imports'}, axis = 1)

# Select data
expdat = world.merge(expdat, left_on = 'iso_a3', right_on = 'Country Code').rename({'2019': 'exports'}, axis = 1)

data = expdat.merge(impdat[['Country Code', 'imports']], on = 'Country Code').drop(['name', 'iso_a3'], axis = 1)
```

```python
print('Following Countries have no recorded exports for 2019: ')
print([i for i in data.loc[data['exports'].isna()]['Country Name'].unique()])
```

```python
print('Following Countries have no recorded imports for 2019: ')
print([i for i in data.loc[data['imports'].isna()]['Country Name']])
```

```python
# Fill missing records
data[['imports', 'exports']] = data[['imports', 'exports']].fillna(0)

# Calculate overall
data['net'] = data['exports'] - data['imports']
```

```python
# Visualize GPD data
data.plot(column = 'net', figsize = [50, 20], legend = True)
plt.axis('off'),
plt.title('World Map Based On Net Export Traffic');
```
